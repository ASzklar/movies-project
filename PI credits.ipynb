### Genero DFs

```python
df_credits = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/credits.csv')
df_credits_copy = df_credits.copy()
```

```python
# Ejecuto este comando para poder visualizar todas las columnas cuando pido un .head() por ejemplo

pd.set_option('display.max_columns', None)
```

```python
df_credits_copy.shape
```

```python
# Elimino los registros duplicados

df_credits_copy.drop_duplicates(inplace=True)
```

```python
df_credits_copy.shape
```

```python
# Creo la columna 'names_cast' con la información de la columna 'cast'

def convert_to_dict(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return None

df_credits_copy['cast'] = df_credits_copy['cast'].apply(convert_to_dict)

def name(x):
    if isinstance(x, list) and len(x) > 0:
        return ', '.join([cast['name'] for cast in x])
    else:
        return None

df_credits_copy['names_cast'] = df_credits_copy['cast'].apply(name)
```

```python
# Creo la columna 'director' con la información de la columna 'crew'

df_credits_copy['crew'] = df_credits_copy['crew'].apply(ast.literal_eval)

def extraer_director(crew):
    for entry in crew:
        if entry['job'] == 'Director':
            return entry['name']
    return None

df_credits_copy['director'] = df_credits_copy['crew'].apply(extraer_director)
```

```python
df_credits_copy.drop(['cast', 'crew'], axis=1, inplace=True)
df_credits_copy.rename(columns={'names_cast': 'cast'})
```

```python
df_credits_copy.to_csv('/content/drive/MyDrive/Colab Notebooks/credits_limpio.csv', index=False)
```
